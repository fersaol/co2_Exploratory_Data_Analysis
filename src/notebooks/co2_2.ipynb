Vemos que el dataframe anterior tiene los datos de longitud y latitud promediados y por eso buscamos uno alternativo encontrandolo en una url de google:

```python
location = pd.read_html("https://developers.google.com/public-data/docs/canonical/countries_csv")
type(location)
```

Vemos que es de tipo lista porque al intentar abrirlo como dataframe nos decÃ­a que no era de dos dimensiones, por lo que lo pasamos a un array de numpy

```python
location = np.array(location)
```

vemos que no es de dos dimensiones sino que tiene cuatro, por lo que le hacemos un reshape

```python
location.shape
```

```python
location = location.reshape(245,4)
```

Convertimos el array de numpy en dataframe y renombramos las columnas:

```python
df_locations = pd.DataFrame(location)
df_locations = df_locations.rename(columns={0:"CODE",1:"latitude",2:"longitude",3:"Country"})
df_locations.head()
```

Eliminamos las columnas antiguas:

```python
df = df.drop(columns=["longitude","latitude"])
df.info()
```

juntamos ambos dataframes

```python
df = pd.merge(df,df_locations,on="Country",how="outer")
df.head()
```
